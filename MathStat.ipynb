## Задание 2
Смоделируем центральную предельную теорему, которая звучит следующим образом.
Положим $$S_{n} = \sum_{k=1}^{k=n} \xi_{k}$$  Сумма независимых одинаково распределенных случайных величин
Матожидание: $$ES_{n}$$
Дисперсия: $$DS_{n}$$
Тогда при ненулевой дисперсии и конечном матожидании: $$\frac{S_{n} - ES_{n}}{\sqrt{DS_{n}}} \Rightarrow{N(0,1)} **$$

для каждого n создадим n выборок размера k и эмпирически оценим распределение

```python
#Задание 2
def get(n, table):
    mean, var = lognorm.stats(s = sigma, loc = mu, moments='mv')
    x = (np.sum(table[:n]) - n * mean)/(sqrt(n*var))
    return x
```

```python
gen = [lognorm.rvs(s = sigma, loc = mu, size = 10000) for i in range (10000)]
```

```python
def CLT(n,k,old):
    res = []
    for val in range(k):
        x = get(n, old[val])
        res.append(x)
    t = np.linspace(-3,3,60)
    plt.hist(res, density = True, bins = int(sqrt(k)), color = 'red', label = 'Empirical N(0,1)')
    x = np.linspace(norm.ppf(0.001), norm.ppf(0.999), 100)
    plt.plot(x, norm.pdf(x), 'b-', lw=5, alpha=0.6, label='norm pdf')
    plt.legend()
```

Покажем, что с ростом n разность между плотностью и горизонтальными отрезками гистограммы уменьшается.

Нашей выборке и нормальному распределению соответствуют эмпирическая функция распределения и функция распределения нормальной случайной величины.

```python
def RetCLT(n,k,old):
    res = []
    for val in range(k):
        x = get(n, old[val])
        res.append(x)
    return res
```

```python
def calculate_supremum(res, i, n):
    return i/n - norm.cdf(res[i - 1])
```

```python
res = sorted(res)
```

Рассчитаем "расстояние" между двумя функциями и покажем, как оно зависит от n. За расстояние возьмем супремум разности между двумя значениями функции.

```python
def get_supremums(n, res):
    exit = []
    for i in range(1, n + 1):
        exit.append(abs(calculate_supremum(res, i, n)))
    return exit
```

```python
lst = get_supremums(5000, res)
```

```python
plt.plot(np.linspace(1,5000,5000), lst, color = 'red', label = 'Error')
plt.xlabel("Количество испытаний")
plt.ylabel("Величина ошибки")
plt.legend()
```

Отсюда видим, что "расстояние" между функциями уменьшается(выполняется теорема Гливенко-Кантелли и эмпирическая функция все ближе к непрерывной функции распределения нормальной случайной величины). Это так же означает, что гистограмма с ростом n все больше становится похожа на график плотности нормального распределения.

Посмотрим асимптотику при помощи неравенства Бэрри-Эссеена

### Ядерная оценка плотности

Воспользуемся ядерной оценкой плотности, для этого каждой точке выборки сопоставим гауссовское ядро и усредним:
    $$f_{emp}(x) = \frac{\sum_{i = 0}^{i = n} f(x)}{n}$$, где f(x) - гауссовское ядро:
    $$f(X_{i}) = \frac{1}{\sqrt{2\pi}}e^{-\frac{(x-X_{i})^2}{2}}$$
Более "адекватная" оценка плотности может быть достигнута ядром Епанечникова.
Пример описан на языке R в приложении 1.

Асимптотика Бэрри-Эссеена

```python
def Berry_Esseen(n, sigma, mean, C = 0.4785):
    vr = sqrt(sigma)
    return C*skew/sqrt(n)
```

## Задание 3
Моделирование пуассоновского случайного процесса

Необходимо решить задачу: Пусть диспетчеру поступают звонки согласно Пуассоновскому процессу с интенсивностью l звонков в час. В 3 часа дня диспетчер делает перерыв на чай в течении 0.5 часа. В этот промежуток времени поступил ровно один звонок. Сколько в среднем времени пройдет от этого звонка до последующего.

Смоделируем задачу так: воспользуемся первым определением Пуассоновского процесса: Первый звонок есть реализация 
случайной величины с экспоненциальным распределением. По принципу отсутствия последействия получим реализацию еще одной случайной величины с экспоненциальным распределением. Первая величина должна быть в пределах от 0 до 0.5 часа, а сумма первой и второй: после 0,5 часа.


```python
def build_Poisson_process(l, n):
    result = []
    counter = 0
    while(counter < n):
        rv1 = expon.rvs(scale = 1/l)
        if(rv1 < 0.5):
            rv2 = expon.rvs(scale = 1/l)
            if(rv1 + rv2 > 0.5):
                result.append(rv2)
                counter += 1
    return result
```

```python
result = []
vars = np.linspace(1,10,45)
for i in vars:
    t = build_Poisson_process(i,500)
    result.append(np.sum(np.array(t))/500)
```

```python
plt.plot(vars, result, color = 'red', label = 'Empirical time')
plt.plot(vars, [1/4 + 1/i for i in vars], color = 'blue', label = 'Theoretical time')
plt.xlabel("lambda")
plt.ylabel("average time")
plt.legend()
```

Отсюда видим, что численно получили среднее время ожидания звонка:
$$ET = \frac{1}{4} + \frac{1}{l}$$

Обернем выполненные действия и посмотрим, что будет происходит с ростом количества испытаний для каждого параметра L

```python
def Dependence(k,l,m,n):
    result = []
    vars = np.linspace(k,l,m)
    for i in vars:
        t = build_Poisson_process(i,n)
        result.append(np.sum(np.array(t))/n)
    plt.plot(vars, result, color = 'red', label = 'Empirical time')
    r = np.linspace(k,l,1000)
    plt.plot(r, [1/4 + 1/i for i in r], color = 'blue', label = 'Theoretical time')
    plt.xlabel("lambda")
    plt.ylabel("average time")
    plt.legend()
```
